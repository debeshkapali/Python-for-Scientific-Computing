## Part 1: Basic Python

```python
# Simple list and loop
temperatures = [20, 22, 19, 23, 21, 24, 20]

print("Daily temperatures:", temperatures)

# Calculate average using a loop
total = 0
for temp in temperatures:
    total += temp

average = total / len(temperatures)
print(f"Average temperature: {average:.1f}°C")
```

```python
# Using conditionals
for i, temp in enumerate(temperatures):
    if temp > 22:
        print(f"Day {i+1}: {temp}°C - Warm day!")
    else:
        print(f"Day {i+1}: {temp}°C")
```

## Part 2: NumPy Basics

```python
import numpy as np

# Create array from list
temps_array = np.array(temperatures)
print("Array:", temps_array)
print("Type:", type(temps_array))
```

```python
# NumPy makes calculations easier
print("Mean:", temps_array.mean())
print("Max:", temps_array.max())
print("Min:", temps_array.min())
print("Std:", temps_array.std())
```

```python
# Simple array operations
temps_fahrenheit = temps_array * 9/5 + 32
print("Temperatures in Fahrenheit:", temps_fahrenheit)
```

## Part 3: Pandas for Data

```python
import pandas as pd

# Create a simple DataFrame
data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temperature': [20, 22, 19, 23, 21, 24, 20],
    'Humidity': [65, 70, 80, 60, 68, 55, 72]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Basic operations
print("Average temperature:", df['Temperature'].mean())
print("Average humidity:", df['Humidity'].mean())
```

```python
# Filter data
warm_days = df[df['Temperature'] > 21]
print("Warm days (>21°C):")
print(warm_days)
```

## Part 4: Simple Plotting

```python
import matplotlib.pyplot as plt

# Simple line plot
plt.figure(figsize=(8, 4))
plt.plot(df['Day'], df['Temperature'], marker='o')
plt.title('Temperature Over the Week')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(8, 4))
plt.bar(df['Day'], df['Humidity'])
plt.title('Humidity Levels')
plt.xlabel('Day')
plt.ylabel('Humidity (%)')
plt.show()
```

## Part 5: Reading and Writing Files

```python
# Save DataFrame to CSV
df.to_csv('weather_data.csv', index=False)
print("Data saved to weather_data.csv")
```

```python
# Read it back
df_loaded = pd.read_csv('weather_data.csv')
print("Data loaded:")
print(df_loaded)
```
